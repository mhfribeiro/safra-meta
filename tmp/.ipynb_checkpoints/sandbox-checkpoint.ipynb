```python
import pandas as pd

df = pd.read_excel('cervejas.xlsx')

numeric_data = df._get_numeric_data()

maxima = numeric_data.max()
minima = numeric_data.min()

data_range = maxima - minima
data_range[data_range == 0] = 1.0

numeric_data = (numeric_data - minima) / data_range

df[numeric_data.columns] = numeric_data

df
```

```python
numeric_data = df._get_numeric_data()
numeric_data = numeric_data * data_range + minima

df[numeric_data.columns] = numeric_data

df
```

```python
import numpy as np
import scipy.stats as sts
limQ1 = 25
limQ3 = 75
c=1.5


q1 = np.percentile(numeric_data, limQ1, axis=0)
q3 = np.percentile(numeric_data, limQ3, axis=0)
iqr = sts.iqr(numeric_data, axis=0)

keep = []
inf = []
sup = []

for i in range(len(numeric_data)):
    d = numeric_data.loc[numeric_data.index[i]]
    test_inf = d < q1 - c * iqr
    if test_inf.any():
        inf.append(i)
    else:
        test_sup = d > q3 + c * iqr
        if test_sup.any():
            sup.append(i)
        else:
            keep.append(i)

print(inf)
outliers_inf = df.loc[df.index[inf]]

outliers_inf
```

```python
print(sup)
outliers_sup = df.loc[df.index[sup]]
outliers_sup
```

```python
df.drop(inf + sup, inplace=True)

df
```
